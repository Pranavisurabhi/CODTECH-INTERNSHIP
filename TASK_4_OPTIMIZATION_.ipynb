```python
import pulp
```

Maximize:
Profit = 20x + 30y

📝 Constraints
2x + y ≤ 100
x + 3y ≤ 90

```python
# Create optimization problem
model = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# Decision variables
x = pulp.LpVariable('Product_A', lowBound=0)
y = pulp.LpVariable('Product_B', lowBound=0)

# Objective function (Maximize profit)
model += 20 * x + 30 * y

# Constraints
model += 2 * x + y <= 100
model += x + 3 * y <= 90

# Solve the problem
model.solve()

# Results
print("Status:", pulp.LpStatus[model.status])
print("Optimal Product A:", x.value())
print("Optimal Product B:", y.value())
print("Maximum Profit:", pulp.value(model.objective))
```

```python
import matplotlib.pyplot as plt

products = ['Product A', 'Product B']
values = [x.value(), y.value()]

plt.bar(products, values)
plt.title("Optimal Production")
plt.xlabel("Products")
plt.ylabel("Quantity")
plt.show()
```

The optimal number of products is obtained using linear programming, maximizing profit efficiently.
